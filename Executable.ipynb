```python
import gurobipy as gp
from gurobipy import Model, GRB, quicksum
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from geopy import distance #si les arroja error, coloquen en anaconda prompt conda install -c conda-forge geopy

ll = int(input("Insert matrix size: "))

Q=200
V=10
df=pd.read_excel("Customers data/Med1.xlsx")
df.set_index('Cliente',inplace=True)
nodos=[a for a in df.index]
clientes=[a for a in nodos if a!=0]
vehiculos=[i for i in range(1,V+1)]
df

import folium
from folium import features
from folium.plugins import Draw

mapa = folium.Map(location=[df['lat'].mean(),df['lng'].mean()],zoom_start=14)


fg=folium.FeatureGroup()  
for n in nodos:
    
    customer_id = df['nro_cliente'][n]
    customer_name = df['customer_id'][n] # Parametro de folium.Marker: muestra data cuando pasas por encima de un Marker
    location = df['lat'][n],df['lng'][n] # Coordenadas del punto n
    
    fg.add_child(folium.CircleMarker(location=location, 
                                     radius=3,
                                     color='#ff0000',
                                     fill = True,
                                     fill_color='#ff0000'))
    
    fg.add_child(folium.Marker(location=location,
                               popup=folium.Popup(str(customer_id) + " " + customer_name),
                               tooltip=str(customer_id),
                               icon=folium.Icon(color='green',
                                            icon_color='white',
                                            icon='info-sign')))

mapa.add_child(fg)

len(df.index)-1  #  = 20

polylines = []

for i in range(len(df.index) - 1):
    start_point = df['lat'][i],df['lng'][i]
    end_point = df['lat'][i + 1],df['lng'][i + 1]
    
    polyline = folium.PolyLine(locations=[start_point,end_point],color='blue')
    
    polylines.append(polyline)
    
for polyline in polylines:
    polyline.add_to(mapa)

mapa
```
